```python
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
def ler_json(arq_json):
    with open(arq_json, 'r', encoding='utf8') as f:
        return json.load(f)
```

```python
treino_unet_base1 = '/home/anatielsantos/mestrado/datasets/dissertacao/final_tests_dis/original_unet_test2_exp1_2.json'
treino_unet_base2 = '/home/anatielsantos/mestrado/datasets/dissertacao/final_tests_dis/original_unet_test2_exp2_2.json'

treino_gan_base1 = '/home/anatielsantos/mestrado/datasets/dissertacao/final_tests_dis/gan_test2_exp1_2_history_150epc.json'
treino_gan_base2 = '/home/anatielsantos/mestrado/datasets/dissertacao/final_tests_dis/gan_test2_exp2_2_history_150epc.json'
```

```python
def plot_json_unet(json, title):
    history = ler_json(json)
    plt.plot(history[list(history.keys())[1]].values())
    plt.plot(history[list(history.keys())[3]].values())
    # plt.plot(history['g_l1'])
    plt.title(title)
    plt.ylabel('Dice')
    plt.xlabel('Épocas')
    plt.legend(['Dice Treino', 'Dice Validação'], loc='lower right')
    # save plot to file
    # plt.savefig(path_plot+'gan_test2_exp1_1_plot_150epc.png')
    plt.show()
    
def plot_json_gan(json, title):
    history = ler_json(json)
    plt.plot(history[list(history.keys())[4]].values())
    plt.plot(history[list(history.keys())[5]].values())
    # plt.plot(history['g_l1'])
    plt.title(title)
    plt.ylabel('Dice')
    plt.xlabel('Épocas')
    plt.legend(['Dice Treino', 'Dice Validação'], loc='lower right')
    # save plot to file
    # plt.savefig(path_plot+'gan_test2_exp1_1_plot_150epc.png')
    plt.show()
```

```python
plot_json_unet(treino_unet_base1, 'Experimento com a Base 1')
```

```python
plot_json_unet(treino_unet_base2, 'Experimento com a Base 2')
```

```python
plot_json_gan(treino_gan_base1, 'Experimento com a Base 1')
```

```python
plot_json_gan(treino_gan_base2, 'Experimento com a Base 2')
```
